## Importing Required libraries

```python
import pandas as pd
from chembl_webresource_client.new_client import new_client 
```

### Searching for Target protein

#### Target search for Malaria

```python
#Target search for malaria
target =new_client.target
target_query=target.search('Dengue')
targets=pd.DataFrame.from_dict(target_query)
targets.head(6)
```

### Select and retrive bioactivity data for dengue virus type-2(6th entry)

```python
selected_target=targets.target_chembl_id[5]
selected_target
```

#### Retriving the bioactivity data for dengue virus type 2 (CHEMBL5980)

```python
activity=new_client.activity
res=activity.filter(target_chembl_id=selected_target).filter(standard_type="IC50")
```

```python
df=pd.DataFrame.from_dict(res)
```

```python
df.head(3)
```

```python
df.type.unique()
```

##### Finally we will save the resulting bioactivity data to CSV file bioactivity_data.csv

```python
df.to_csv('bioactivity_data2.csv',index=False)
```

#### Handiling Missing data

```python
df2=df[df.standard_value.notna()]

df2
```

##### Data pre-processing of the bioactivity data

```python
bioactivity_class=[]
for i in df2.standard_value:
    if float(i)>=10000:
        bioactivity_class.append("inactive")
    elif float(i)<=1000:
        bioactivity_class.append("active")
    else:
        bioactivity_class.append("Intermediate")
```

###### Iterating molecule_chembl_id to a list

```python
df2.molecule_chembl_id
    
```

```python
selection=['molecule_chembl_id','canonical_smiles','standard_value']
df3=df2[selection]
df3
```

```python
df3['bioactivity_class']=(pd.Series(bioactivity_class))
df3
```

```python
df3=df3[df3.bioactivity_class.notna()]
df3
```

```python
df3.to_csv('bioactivity_preprocessed_data3.csv',index=False)
```
